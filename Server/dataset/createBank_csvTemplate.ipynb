```python
template = {
    "category" : [],
    "prestation" : [],
    "type" : [],
    "value" : []
}
```

```python
template["category"] = []

for category in categories.keys():
  for i in range(len(categories[category].keys())*len(types)):
    template["category"].append(category)

template["category"]
```

```python
template["prestation"] = []

for category in categories.keys():
  for prestation in categories[category].keys():
    for j in range(len(types)):
      template["prestation"].append(prestation)

template["prestation"]
```

```python
template["type"] = []

for category in categories.keys():
  for prestation in categories[category].keys():
    for type in types:
      template["type"].append(type)

template["type"]
```

```python
template["value"] = ["" for _ in template["category"]]
```
